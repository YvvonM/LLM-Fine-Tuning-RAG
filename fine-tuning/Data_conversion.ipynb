```python
#importing fasttext library
import fasttext as ft
import json
import pandas as pd
import re
```

### Importing the dataset

```python
file_path = "/home/yvvon_majala/llm-fast-api/Beton_Rentals_🇪🇹.json"

with open(file_path, 'r') as file:
    # Load the JSON data from the file
    data1 = json.load(file)
   
            
```

```python
print(data1)
```

```python
# Extract id, date, and message data
messages_data = []
for message_info in data1.get("messages", []):
    message_data = {
        "ID": message_info.get("id", ""),
        "Date": message_info.get("date", ""),
        "Message": message_info.get("message", ""),
        "label": message_info.get("label", "")
    }
    messages_data.append(message_data)

# Create a DataFrame
df = pd.DataFrame(messages_data)

# Print the DataFrame
print(df.head())
```
