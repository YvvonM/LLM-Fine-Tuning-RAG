#### Import modules

```python

import pandas as pd
from src.util import Util
util = Util()
```

#### Read parsed data

```python
df = pd.read_csv("/home/yvvon_majala/llm-fast-api/data/shuffled.csv")
df.head()
```

```python
df.shape
```

#### Remove null values

```python
df = df.dropna()
df.head()
```

```python
df.shape
```

#### Remove new line

```python
df = df.replace('\n', ' ', regex=True)
df.head()
```

#### Extract and remove hasthags

```python
# Extract hashtags
df['hashtags'] = df['Message'].apply(lambda x: util.extract_hashtags(x))
df.head()
```

```python
# Remove hashtags from text
df['Message'] = df['Message'].str.replace(r'\#\w+', '', regex=True)
df.head()
```

#### Extract emojis

```python
# Extract emojis using regex
df['emojis'] = df['Message'].apply(util.extract_emojis)
df.tail()
```

#### Remove emojis from text

#### Replace letters:
* ['·àê', '·àë', '·àí', '·àì', '·àî', '·àñ'] with ['·àÄ', '·àÅ', '·àÇ', '·àÉ', '·àÑ', '·àÖ', '·àÜ']
* ['·äÄ', '·äÅ', '·äÇ', '·äÉ', '·äÑ', '·äÖ', '·äÜ'] with ['·àÄ', '·àÅ', '·àÇ', '·àÉ', '·àÑ', '·àÖ', '·àÜ']
* ['·à†', '·à°', '·à¢', '·à£', '·à§', '·à¶', '·à¶', '·àß'] with ['·à∞, '·à±', '·à≤', '·à≥', '·à¥', '·àµ', '·à∂', '·à∑']
* ['·ãê', '·ãë', '·ãí', '·ãì', '·ãî', '·ãï', '·ãñ'] with ['·ä†', '·ä°', '·ä¢', '·ä£', '·ä§', '·ä•', '·ä¶']
* ['·å∏', '·åπ', '·å∫', '·åª', '·åº', '·åΩ', '·åæ'] with ['·çÄ', '·çÅ', '·çÇ', '·çÉ', '·çÑ', '·çÖ', '·çÜ']

```python
letters = [
  [['·àê', '·àë', '·àí', '·àì', '·àî', '·àñ'], ['·àÄ', '·àÅ', '·àÇ', '·àÉ', '·àÑ', '·àÖ', '·àÜ']],
  [['·äÄ', '·äÅ', '·äÇ', '·äÉ', '·äÑ', '·äÖ', '·äÜ'], ['·àÄ', '·àÅ', '·àÇ', '·àÉ', '·àÑ', '·àÖ', '·àÜ']],
  [['·à†', '·à°', '·à¢', '·à£', '·à§', '·à¶', '·à¶', '·àß'], ['·à∞', '·à±', '·à≤', '·à≥', '·à¥', '·àµ', '·à∂', '·à∑']],
  [['·ãê', '·ãë', '·ãí', '·ãì', '·ãî', '·ãï', '·ãñ'], ['·ä†', '·ä°', '·ä¢', '·ä£', '·ä§', '·ä•', '·ä¶']],
  [['·å∏', '·åπ', '·å∫', '·åª', '·åº', '·åΩ', '·åæ'], ['·çÄ', '·çÅ', '·çÇ', '·çÉ', '·çÑ', '·çÖ', '·çÜ']]
]
for letter in letters:
  for i in range(len(letter[0])):
    df['Message'] = df['Message'].str.replace(letter[0][i], letter[1][i])
    
```

#### Extract symbols

```python
df['symbols'] = df['Message'].apply(util.extract_symbols)
df.tail()
```

```python
df['Message'] = df['Message'].apply(util.remove_symbols)
df.tail()
```

#### Extract Links

```python
df['links'] = df['Message'].apply(util.extract_urls)
df.tail()
```

#### Remove links 

```python
df['Message'] = df['Message'].str.replace(util.url_pattern, '', regex=True).str.strip()
df.tail()
```

#### Extract mentions

```python
df['mentions'] = df['Message'].apply(util.extract_mentions)
df.tail()
```

```python
df['Message'] = df['Message'].str.replace(util.mention_pattern, '', regex=True).str.strip()
df.tail()
```

#### Remove extra spaces

```python
df['Message'] = df['Message'].str.replace('\s+', ' ', regex=True).str.strip()
```

```python
df['Message'] = df['Message'].replace(r'!+', '!', regex=True)
df['Message'] = df['Message'].replace(r'\.+', '', regex=True)
```
