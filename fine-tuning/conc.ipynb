### Importing libraries

```python
#data manipulation libraries
import pandas as pd
import numpy as np

```

### Importing the dataset

```python
df1 = pd.read_csv('/home/yvvon_majala/llm-fast-api/data/benton_rentals.csv')
df1.head()
```

```python
#dropping unnamed column
df1.drop(['Unnamed: 0'], axis = 1, inplace = True)
df1.tail()
```

```python
#dropping unnamed column
df1.drop(['ID'], axis = 1, inplace = True)
df1.tail()
```

```python
#loading the second dataset
df2 = pd.read_csv("/home/yvvon_majala/llm-fast-api/data/gojo.csv")
df2.head()
```

```python
#dropping unnamed column
df2.drop(['Unnamed: 0'], axis = 1, inplace = True)
df2.tail()
```

### Concatinating the dataset

```python
#concatinating the two datasets
df3 = pd.concat([df1,df2], axis = 0)
df3.shape
```

### Shuffling the dataset rows

```python
# shuffle the DataFrame rows
df3 = df3.sample(frac = 1)
df3.head()
```

```python
#saving the dataset
df3.to_csv("/home/yvvon_majala//llm-fast-api/data/shuffled.csv")
```
